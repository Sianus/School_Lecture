# Visualizing Errors

## Basic Errorbars

Matplotlib function call - ``plt.errorbar()``

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k'); # yerr = err y축(즉, dy를 세로축으로 사용하겠다)
```

``fmt``는 line과 point 모습을 결정하는 속성이다. 

아래 예와 같이 다른 속성도 볼 수 있다.

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```
