### 가정
다음은 100명 학생에 관한 정보를 담고 있는 변수들이다. 
 - sname : 학생 이름
 - noclass : 반 번호
 - gender : 성별
 - height : 키
 - weight : 몸무게
 
아래 코드는 수정없이 사용하면 된다.

```python
### 지우지 마시오.
import numpy as np
import pandas as pd
rng=np.random.RandomState(123)

sname=['s'+format(i, '03d') for i in range(100)]
classno=rng.randint(1, 6, 100)
gender=['female']*50 + ['male']*50
height=rng.randint(160, 190, 100)
weight=rng.randint(50, 80, 100)
```

### 문제1
위에서 생성한 NumPy array들인 sname, classno, gender, height, weight를 column으로 사용하는 DataFrame 변수 ``student``를 생성하시오. 

- index는 implicit index (자동으로 생성되는 0이상 정수)
- i번째 학생의 정보는 sname[i], classno[i], gender[i], height[i], weight[i]

```python
### 답안을 작성하시오. 
student = pd.DataFrame({'sname':sname, 'classno':classno, 'gender':gender, 'height':height, 'weight':weight})
student
```

```python
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
student
```

### 문제2
위 ``student`` 변수의 ``classno``는 1반부터 5반까지 값이 들어있다. 이 때 1반부터 5반까지의 담임교사 이름 각각 Kim, Lee, Park, Choi, Yun이라고 하자. 

반 번호(classno), 담임교사 이름(tname), 반 학생 수(size)의 순서대로 column으로 가지는 DataFrame 변수 ``teacher``를 생성하시오.  
- index는 implicit index (자동으로 생성되는 0이상 정수)
- size는 ``student`` 변수에서 반 별 학생수를 헤아릴 것

```python
### 답안을 작성하시오. 
tname = ['Kim', 'Lee', 'Park','Choi','Yun']
size = student.groupby('classno')['classno'].count()
teacher = pd.DataFrame({'tname':tname, 'size':size}).reset_index('classno')
teacher
```

```python
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
teacher
```

### 문제3

```python
### 답안을 작성하시오. 
info = pd.merge(student, teacher)
info
```

### 문제6
반별로 height, weight, MBI의 평균값을 구하여 아래 결과와 같이 dataframe 변수 ``info_avg``에 저장하시오.
 - ``info``와 ``info_sname`` 중 어떤 변수를 사용하여도 좋다.
 - ``info_avg`` 에는 ``size``가 없어야 한다.

```python
### 답안을 작성하시오. 
info_avg = info.groupby('classno').mean().drop('size', axis=1)
info_avg
```

### 문제7
pivot_table을 사용하여 반별, 성별  MBI 평균값을 구한 결과를 ``info_pivot`` 변수에 저장하시오. 
- 행은 classno, 열은 geneder이다.

```python
### 답안을 작성하시오. 
info_pivot = info.groupby(['classno', 'gender'])['MBI'].aggregate('mean').unstack()
info_pivot
```

### 문제8
전체 학생들 중 MBI값이 전체 MBI 중위값(median)보다 큰 학생들만 가리키는  DataFrame 변수 ``info_high``를 생성하시오.

- 단, 학생들 순서는 현재까지 ``info``에 등장하는 순서이다.

```python
### 답안을 작성하시오. 
info_high = info[info.MBI > (info['MBI'].median())]
info_high.head()
```

```python
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
info_high.head()
```

### 문제9
각 학생의 키에서 남학생 혹은 여학생 전체 평균키를 뺀 값을 *height_dev* 제목의 column으로 *info* 변수에 추가하시오.

```python
### 답안을 작성하시오. 
avg = info.groupby('gender').mean()
info=pd.merge(info, avg['height'],on='gender',suffixes=['','_dev'])
info['height_dev'] = info['height'] - info['height_dev']
info
```

```python
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
info
```

### 문제10
*height_dev* 가 0보다 큰 여학생 수 및 남학생 수를 각각 *ct_tall_female* 및 *ct_tall_male* 변수에 저장하시오.

```python
### 답안을 작성하시오. 
ct_tall_female = ((info.gender=='female') & (info.height_dev > 0)).sum()
ct_tall_male = ((info.gender=='male') & (info.height_dev > 0)).sum()
ct_tall_female, ct_tall_male
```

```python
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
ct_tall_female, ct_tall_male
```
