### 가정
- 아래에서 생성된 _score_ 변수는 $50 \times 5$ 크기의 2차원 numpy array이다. 
- 이는 50명 학생의 국어, 영어, 수학, 과학, 사회 점수(1번째부터 5번째 열 순서)를 각각 나타낸다. (0점~100점)

```
### 수정하지 마시오.
import numpy as np
rng=np.random.RandomState(123)
score=rng.randint(20, 101,(50,5))
score
```

### 문제1
가장 마지막 학생의 수학 점수가 55점이 아니라 75점이라고 한다.

이렇게 *score* 배열을 수정하기 위한 코드를 작성하시오.

```
### 답안을 작성하시오. 
score[-1,-3] = 75
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
print(score[-1])
```

### 문제2
채점에 오류가 발견되어 모든 학생들의 영어점수를 2점씩 줄여야 한다고 한다. 이렇게 *score* 배열을 수정하기 위한 코드를 작성하시오.

```
### 답안을 작성하시오. 
score[:, 1] = score[:, 1] -2
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
score[0], score[-1]
```

### 문제3
50명의 추가 학생들의 점수가 다음 _score\_add_ 변수에 주어진다. 이를 _score_ 변수 데이터 끝에 덧붙여 100명의 데이터를 *score*변수에 다시 저장하고자 한다. 이를 위한 코드를 작성하시오.

```
### 수정하지 마시오.
score_add=rng.randint(0, 101,(50,5))
score_add
```

```
### 답안을 작성하시오. 
score = np.concatenate([score,score_add])
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
score[0], score[49], score[50], score[99]
```

### 문제4
과목별 평균점수를 뜻하는 _avg\_subject_ array(1차원)를 구하는 코드를 작성하시오. 

```
### 답안을 작성하시오. 
avg_subject = score.sum(axis=0)/100
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
avg_subject
```

### 문제5
학생별 평균점수를 뜻하는 _avg\_student_ array(1차원)를 구하는 코드를 작성하시오. 

```
### 답안을 작성하시오. 
avg_student = score.sum(axis=1)/5
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
avg_student
```

### 문제6
각 점수가 해당과목의 평균으로부터 얼마나 차이가 나는지(이를 편차라고 한다.)를 나타내는 _score\_dev_ array(2차원)를 생성하시오.
- score_dev의 shape는 score의 shape와 동일하다. 즉 (100, 5)
- 만약 어떤 학생의 국어 점수가 30점이고 국어 평균이 45점이면 그 차이(편차)는 $30-45=-15$ 점이다.

---
참고) _score\_dev_ 의 axis=0 방향 평균값이 어떻게 될 지 생각해 보자.


```
### 답안을 작성하시오. 
score_dev = score-avg_subject
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
score_dev[0], score_dev[-1]
```

### 문제7
전체과목 평균점수를 기준으로 1등을 한 학생과 꼴등을 한 학생의 5개 과목 점수 array를 각각 *student_best*, *student_worst*에 각각 저장하시오. 

```
### 답안을 작성하시오. 
student_best = score[avg_student.argmax()]
student_worst = score[avg_student.argmin()]
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
student_best, student_worst
```

### 문제8
전체 학생의 평균점수를 구하여 _avg\_total_ 변수에 저장하시오. 

```
### 답안을 작성하시오. 
avg_total = score.mean()
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
avg_total
```

### 문제9
전체 평균 점수(_avg\_total_) 보다 높은 평균 점수를 얻은 학생의 수를 구하여 _no\_student\_high_ 변수에 저장하시오.  

```
### 답안을 작성하시오. 
no_student_high = np.count_nonzero(avg_student>avg_total)
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
no_student_high
```

### 문제10
과학점수가 사회점수보다 더 높은 학생 수를 구하여 _no\_student\_sci_ 변수에 저장하시오.
 

```
### 답안을 작성하시오. 
no_student_sci = np.count_nonzero(score[:,3] > score[:, 4])
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
no_student_sci
```
